## AIRLINE PASSENGER

```python
DATA_PATH=r'C:\Users\Ankan\Desktop\Github\Time-Series-Forecasting\data'
```

```python
df = pd.read_csv(os.path.join(DATA_PATH,'international-airline-passengers.csv'))
```

```python
df.head(20)
```

```python
from statsmodels.tsa.filters.hp_filter import hpfilter
```

```python
# Decomposes a signal into cyclical and trend components
cycle, trend = hpfilter(df['n_pass_thousands'], 1600)
```

```python
trend.plot()
```

```python
df['trend'] = trend
df['cycle'] = cycle
```

```python
df[['trend', 'n_pass_thousands']].plot()
```
